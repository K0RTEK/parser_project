```python
import pandas as pd
import numpy as np
import nltk
from nltk.stem.snowball import SnowballStemmer
import re
import os
```

```python
access = pd.read_excel('file.xlsx')
```

```python
access = access[0]
```

```python
text = input('Vvedite tekst: ')
```

```python
text = re.sub(r'^https?://.[\r\n]', '', text, flags=re.MULTILINE)
```

```python
text = re.sub('http://\S+|https://\S+', '', text)
#OR 
text = re.sub('http[s]?://\S+', '', text)
```

```python
split_regex = re.compile(r'[.|!|?|…]')
sentences = filter(lambda t: t, [t.strip() for t in split_regex.split(text)])

sentences_list = []

for sentence in sentences:
    sentences_list.append(sentence)
    
```

```python
stemmer = SnowballStemmer("russian")

def token_and_stem(sentence):
    tokens = [word for sent in nltk.sent_tokenize(sentence) for word in nltk.word_tokenize(sent)]
    filtered_tokens = []
    for token in tokens:
        if re.search('[а-яА-Я]', token):
            filtered_tokens.append(token)
    stems = [stemmer.stem(t) for t in filtered_tokens]
    return stems
```

```python
stopwords = nltk.corpus.stopwords.words('russian')
stopwords.extend(['что', 'это', 'так', 'вот', 'быть', 'как', 'в', 'к', 'на', 'при', 'т.ч', 'если', 'и/или', 'и/', 'ваш', 'или', 'собой' ]) ### Расширение списка
print(stopwords)
```

```python
good_sentences = []
for sentence in sentences_list:
    good_words = []
    for word in token_and_stem(sentence): 
        if word not in stopwords:
            good_words.append(word) 
    good_sentences.append(good_words)
```

```python
itog_sentence = []
for idx, sentence in enumerate(good_sentences):
    if len(sentence) > 5: 
        access_count = 0
        
        print(sentence)
        try: 
            for word in good_sentences[idx]: 
                if word in list(access):
                    all_count = len(sentence)
                    access_count += 1          
            Procentil = access_count/all_count
            if Procentil > 0.35:
                 itog_sentence.append(sentences_list[idx])               
        except:
            None
            
            
```

```python
text = f'Было предложений: {len(sentences_list)}\nСтало предложений: {len(itog_sentence)}\n\n'
for idx, i in enumerate(itog_sentence):
    text += str(idx + 1) + '. ' + i + '.' + '\n'   
    
```
